```python
# Download CIFAR-10 binary if missing
import os, pathlib, tarfile, urllib.request, subprocess, shutil
root = pathlib.Path('/content/Parallel-Programming')
data_dir = root / 'Data' / 'cifar-10-batches-bin'
data_dir.parent.mkdir(parents=True, exist_ok=True)
if not data_dir.exists():
    url = 'https://www.cs.toronto.edu/~kriz/cifar-10-binary.tar.gz'
    tar_path = '/content/cifar-10-binary.tar.gz'
    print('Downloading CIFAR-10 binary...')
    subprocess.run(['wget', '-nv', '-O', tar_path, url], check=True)
    subprocess.run(['tar', '-xf', tar_path, '-C', str(root / 'Data')], check=True)
    print('Done extracting.')
else:
    print('Dataset already present, skip download.')
print('Data dir exists:', data_dir.exists())

```
